# Data Cleaning

```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"E:\Unified Mentor\Projects\Employee Attrition Analysis\Attrition data.csv")
```

```python
data.head()
```

```python
# Drop Irrelevant Columns

data.drop(['EmployeeID', 'EmployeeCount', 'Over18', 'StandardHours'], axis=1, inplace=True)
```

```python
data.shape
```

```python
data.describe()
```

```python
# Checking Null Values

data.isnull().sum()
```

```python
# Calculating Total Zeros

(data == 0).sum()
```

```python
# Replace 0 to nan 

data = data.replace(0, np.nan)
```

```python
# Again checking Null values

data.isnull().sum()
```

```python
data.shape
```

# Dealing with Missing Values

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(5, 2, figsize=(15, 20))

sns.distplot(data["NumCompaniesWorked"], ax=axes[0, 0])
sns.distplot(data["StockOptionLevel"], ax=axes[0, 1])
sns.distplot(data["TotalWorkingYears"], ax=axes[1, 0])
sns.distplot(data["TrainingTimesLastYear"], ax=axes[1, 1])
sns.distplot(data["YearsAtCompany"], ax=axes[2, 0])
sns.distplot(data["YearsSinceLastPromotion"], ax=axes[2, 1])
sns.distplot(data["YearsWithCurrManager"], ax=axes[3, 0])
sns.distplot(data["EnvironmentSatisfaction"], ax=axes[3, 1])
sns.distplot(data["JobSatisfaction"], ax=axes[4, 0])
sns.distplot(data["WorkLifeBalance"], ax=axes[4, 1])

plt.tight_layout()
plt.show()
```

Need to use "Mean" -->>  Normal Distribution ( datas in charts evenly distributed means ) 

Need to use "Median" -->>  left or right sided skewed ( Curve will be in single side(not normally distributed means ) 

```python
# Most of the Plots are Left Side Skewed and few Plots are Right Side Skewed

# So we are Going to Fillna with Median Accordingly
```

```python
# Fill na Using Median

data["NumCompaniesWorked"].fillna(data["NumCompaniesWorked"].median(), inplace=True)
data["StockOptionLevel"].fillna(data["StockOptionLevel"].median(), inplace=True)
data["TotalWorkingYears"].fillna(data["TotalWorkingYears"].median(), inplace=True)
data["TrainingTimesLastYear"].fillna(data["TrainingTimesLastYear"].median(), inplace=True)
data["YearsAtCompany"].fillna(data["YearsAtCompany"].median(), inplace=True)
data["YearsSinceLastPromotion"].fillna(data["YearsSinceLastPromotion"].median(), inplace=True)
data["YearsWithCurrManager"].fillna(data["YearsWithCurrManager"].median(), inplace=True)
data["EnvironmentSatisfaction"].fillna(data["EnvironmentSatisfaction"].median(), inplace=True)
data["JobSatisfaction"].fillna(data["JobSatisfaction"].median(), inplace=True)
data["WorkLifeBalance"].fillna(data["WorkLifeBalance"].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Save the Cleaned data

data.to_csv('E:/Unified Mentor/Projects/Employee_Attrition_Analysis_DataCleaned.csv', index = False)
```

# Step 4 - Correlation Analysis

Identify relationships between features and Attrition

```python
# Convert Attrition to numeric
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
```

```python

# Select only numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=[float, int])

# Calculate the correlation matrix for numeric columns only
corr_matrix = numeric_data.corr()

# Plotting the correlation matrix

plt.figure(figsize=(12, 8))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Observation from Step 4

   TotalWorkingYears and Age: Older employees tend to have more years of work experience, which is expected.

3) Weak or No Correlations:

3) Factors to Focus On:
